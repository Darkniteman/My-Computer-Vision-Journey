```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image=mpimg.imread('triangle_tile.jpeg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
print(image.shape)
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
print(gray.shape)
```

```python
gray=np.float32(gray)
dst=cv2.cornerHarris(gray,2,3,0.04)
plt.imshow(dst,cmap='gray')
```

```python
#Dilate function is used to enlarge bright reigon
dst=cv2.dilate(dst,None)
plt.imshow(dst)
```

```python
threshold= 0.0001*dst.max()
print(threshold)
```

```python
corner_image = np.copy(image)
#choose the original image not dst as copy for corner image
for i in range(dst.shape[0]):
    for j in range(dst.shape[1]):
        if dst[i, j] > threshold:
            cv2.circle(corner_image, (i, j), 3, (0, 0, 255), 1)

plt.imshow(corner_image)
plt.show()
```
